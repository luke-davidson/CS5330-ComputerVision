```python
#Luke Davidson - ME 5250 - Final Exam Question 3a

# initialize params
L1 = 2
L2 = 3
theta11_deg = 60
theta12_deg = 30
theta11_rad = theta11_deg*np.pi/180
theta12_rad = theta12_deg*np.pi/180
theta21_deg = 167
theta22_deg = -156
theta21_rad = theta21_deg*np.pi/180
theta22_rad = theta22_deg*np.pi/180
omega1 = 1
omega2 = 3

# calculate Jacobians
Jacobian1 = np.around(np.array([[(-L1*np.sin(theta11_rad)-L2*np.sin(theta11_rad+theta12_rad)),-L2*np.sin(theta11_rad+theta12_rad)],
                     [(L1*np.cos(theta11_rad)+L2*np.cos(theta11_rad+theta12_rad)),L2*np.cos(theta11_rad+theta12_rad)]]),decimals=3)
Jacobian2 = np.around(np.array([[(-L1*np.sin(theta21_rad)-L2*np.sin(theta21_rad+theta22_rad)),-L2*np.sin(theta21_rad+theta22_rad)],
                     [(L1*np.cos(theta21_rad)+L2*np.cos(theta21_rad+theta22_rad)),L2*np.cos(theta21_rad+theta22_rad)]]), decimals=3)
print('-----------------------------------------------')
print('                   Jacobians\n')
print('Jacobian for theta1 = {} and theta2 = {}:\n{}\n'.format(theta11_deg, theta12_deg, Jacobian1))
print('Jacobian for theta1 = {} and theta2 = {}:\n{}'.format(theta21_deg, theta22_deg, Jacobian2))
print('-----------------------------------------------\n')

# initialize angular velocity vector
angular_vel = np.array([[omega1],[omega2]])

# calculate linear velocities v = Jw
linear_vel1 = np.around(Jacobian1 @ angular_vel,decimals=2)
linear_vel2 = np.around(Jacobian2 @ angular_vel,decimals=2)

# initialize force vectors
force11 = 30
force12 = 20
force1 = np.array([[force11],[force12]])
force21 = 30
force22 = -20
force2 = np.array([[force21],[force22]])

# calculate torques tau = J.T * F
torque1 = np.around(Jacobian1.T @ force1,decimals=2)
torque2 = np.around(Jacobian2.T @ force2,decimals=2)
resulting_torque1 = np.around((torque1[0,0]-torque1[1,0]),decimals=2)
resulting_torque2 = np.around((torque2[0,0]-torque2[1,0]),decimals=2)

# print results
print('-----------------------------------------------')
print('                   Case 1\n')
print('For theta1 = {}, theta2 = {}, Force = ({},{}):'.format(theta11_deg, theta12_deg, force11, force12))
print('\tLinear X Velocity: {} m/s'.format(linear_vel1[0,0]))
print('\tLinear Y Velocity: {} m/s'.format(linear_vel1[1,0]))
print('\tResulting joint torque: {} N-m'.format(resulting_torque1))
print('-----------------------------------------------\n')
print('-----------------------------------------------')
print('                   Case 2\n')
print('For theta1 = {}, theta2 = {}, Force = ({},{}):'.format(theta21_deg, theta22_deg, force21, force22))
print('\tLinear X Velocity: {} m/s'.format(linear_vel2[0,0]))
print('\tLinear Y Velocity: {} m/s'.format(linear_vel2[1,0]))
print('\tResulting joint torque: {} N-m'.format(resulting_torque2))
print('-----------------------------------------------\n')

# Conclusion
print('Conclusion: Comparing the initial angles (theta1 and theta2), it is seen that due to starting in different')
print('\torientations, the resulting torques from the differing forces act in opposite directions. Because')
print('\tthe resulting torque from force 1 is negative, it is acting in the clockwise direction. Vice versa, ')
print('\tdue to the resulting torque from force 2 being positive, it is acting in the counter-clockwise direction.')
```

```python
#Luke Davidson - ME 5250 - Final Exam Question 3c

# initialize params
L1 = 2
L2 = 3
m1 = 1
m2 = 2
theta1_deg = 60
theta2_deg = 30
theta1_rad = theta11_deg*np.pi/180
theta2_rad = theta12_deg*np.pi/180
omega1 = 1
omega2 = 10
alpha1 = 1
alpha2 = 1.5
grav = 9.8

# initialize matrices M(Θ), Θ.., V(Θ,Θ.), G(Θ)
M = np.empty((2,2))
alpha = np.empty((2,1))
V = np.empty((2,1))
G = np.empty((2,1))

# assign values
    # M
M[0,0] = (L2**2)*m2+2*L1*L2*m2*np.cos(theta2_rad)+(L1**2)*(m1+m2)
M[0,1] = (L2**2)*m2+L1*L2*m2*np.cos(theta2_rad)
M[1,0] = (L2**2)*m2+L1*L2*m2*np.cos(theta2_rad)
M[1,1] = (L2**2)*m2
    # alpha
alpha[0,0] = alpha1
alpha[1,0] = alpha2
    # V
V[0,0] = -m2*L1*L2*np.sin(theta2_rad)*(omega2**2)-2*m2*L1*L2*np.sin(theta2_rad)*omega1*omega2
V[1,0] = m2*L1*L2*np.sin(theta2_rad)*(omega1**2)
    # G
G[0,0] = m2*L2*grav*np.cos(theta1_rad+theta2_rad)+(m1+m2)*L1*grav*np.cos(theta1_rad)
G[1,0] = m2*L2*grav*np.cos(theta1_rad+theta2_rad)

# calculate resulting tau equation components
M_comp = M @ alpha
V_comp = V
G_comp = G
M1 = M[0,0]
M2 = M[1,0]
b1 = V[0,0]
b2 = V[1,0]
k1 = G[0,0]
k2 = G[1,0]
MT = M1 - M2    # total M
bT = b1 - b2    # total b
kT = k1 - k2    # total k

# create A and B matrices
A1 = np.array([[0,1],[-k1/M1,-b1/M1]])
B1 = np.array([[0],[1/M1]])
A2 = np.array([[0,1],[-k2/M2,-b2/M2]])
B2 = np.array([[0],[1/M2]])
AT = np.array([[0,1],[-kT/MT,-bT/MT]])    # total A
BT = np.array([[0],[1/MT]])    # total B

# calculate initial pole components
aa1 = 1
bb1 = -A1[1,1]
cc1 = -A1[1,0]
aa2 = 1
bb2 = -A2[1,1]
cc2 = -A2[1,0]
aaT = 1    # total aa = 2
bbT = -AT[1,1]    # total bb = 4
ccT = -AT[1,0]    # total cc = 2

# calculate roots before stabilization
pole11 = np.around((-bb1 + np.sqrt((bb1**2)-4*aa1*cc1))/(2*aa1),decimals=2)
pole12 = np.around((-bb1 - np.sqrt((bb1**2)-4*aa1*cc1))/(2*aa1),decimals=2)
pole21 = np.around((-bb2 + np.sqrt((bb2**2)-4*aa2*cc2))/(2*aa2),decimals=2)
pole22 = np.around((-bb2 - np.sqrt((bb2**2)-4*aa2*cc2))/(2*aa2),decimals=2)
pole1T = np.around((-bbT + np.sqrt((bbT**2)-4*aaT*ccT))/(2*aaT),decimals=2)    # total pole 1
pole2T = np.around((-bbT - np.sqrt((bbT**2)-4*aaT*ccT))/(2*aaT),decimals=2)    # total pole 2

# print results
print('-----------------------------------------------------------------')
print('                   Poles before Controller')
print('                            Eq. 1')
print('Pole 1: {}\nPole 2: {}'.format(pole11, pole12))
print('                            Eq. 2')
print('Pole 1: {}\nPole 2: {}'.format(pole21, pole22))
print('                            Total')
print('Pole 1: {}\nPole 2: {}'.format(pole1T, pole2T))
print('-----------------------------------------------------------------\n')

# initialize new poles
npole11 = -1
npole12 = -1
npole21 = npole11
npole22 = npole12
npole1T = npole11    # total new pole 1
npole2T = npole12    # total new pole 2

# calculate k needed to move to new poles
k11 = np.around((-npole11*npole12 - A1[1,0])/B1[1,0], decimals=2)
k12 = np.around((-A1[1,1] - npole11 - npole12)/B1[1,0], decimals=2)
k21 = np.around((-npole21*npole22 - A2[1,0])/B2[1,0], decimals=2)
k22 = np.around((-A2[1,1] - npole21 - npole22)/B2[1,0], decimals=2)
k1T = np.around((-npole1T*npole2T - AT[1,0])/BT[1,0], decimals=2)    # total k1 needed
k2T = np.around((-AT[1,1] - npole1T - npole2T)/BT[1,0], decimals=2)    # total k2 needed

# print results
print('\n\n-----------------------------------------------------------------')
print('                           K values')
print('                             Eq. 1')
print('K values needed to change poles to {} and {}:'.format(npole11, npole12))
print('\tK1: {}'.format(k11))
print('\tK2: {}'.format(k12))
print('                             Eq. 2')
print('K values needed to change poles to {} and {}:'.format(npole21, npole22))
print('\tK1: {}'.format(k21))
print('\tK2: {}'.format(k22))
print('                             Total')
print('K values needed to change poles to {} and {}:'.format(npole1T, npole2T))
print('\tK1: {}'.format(k1T))
print('\tK2: {}'.format(k2T))
print('-----------------------------------------------------------------')
```

```python
import numpy as np

start = 0
right = 0
total = 0

while True:
    if start == 0:
        play = input('Are you ready to begin? (Yes/No):\t')
        if play == 'No':
            print("Well come back when you're ready.")
    else:
        play = input('Do you want to play again? (Yes/No):\t')
        right = 0
        total = 0
        
    ########## GAME ##########
    if play == 'Yes':
        if start == 0:
            username = input('Player username:\t')
            start += 1
        times = int(input('Hi {}! How many times do you want to roll?:\t'.format(username)))
        
#         sides = int(input('How many sides is your die?:\t'))
        sides = 6
    
        while sides < 4:
            print("A die must have at least 4 sides. Please guess a number greater than 3.")
            sides = int(input('Number of sides:\t'))
        percent_correct = np.around((1/sides)*100,decimals=2)
        print('You have a {}% chance of guessing correctly.'.format(percent_correct))
        for i in range(times):
            print('\t\t\t\t\tRound ',i+1)
            guess = int(input('Guess the die value (1-{}):\t'.format(sides)))
            while guess > sides or guess < 1:
                print('Your die has {} sides, so a {} is not possible. Please guess a number between 1 and {}.'.format(sides,guess,sides))
                guess = int(input('Your new guess:\t'))
            roll = np.random.randint(1,sides+1)
            print('Roll: {}\t\tYour Guess: {}'.format(roll,guess))
            if guess == roll:
                print('Your guess of {} was correct!'.format(guess))
                right += 1
            else:
                print('Your guess of {} was not correct.'.format(guess))
            total += 1
        percent = np.around((right/total)*100,decimals=2)
        print('\nFinal score: {}/{}'.format(right, total))
        print('You were correct {}% of the time!'.format(percent))
    ########## GAME ##########
    
    elif play == 'No':
        print('Game over.')
        break
    else:
        print('Invalid input. Please input "Yes" or "No"')
        
```
